````python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.

class Calculator:
    
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def sub(a, b):
        return a - b
    
    @staticmethod
    def mul(a, b):
        return a * b
    
    @staticmethod
    def div(a, b):
        return a / b
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age 
        
    def __repr__(self):
        return f'name: {self.name}, age: {self.age}'
    
    def __gt__(self, target):
        if self.age > target.age:
            return True
        return False
````

````python
import random

# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Pokemon:
    
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * 20
        self.exp = 0
        
    
    def bark(self):
        print(f'{self.name}ì˜ ìš¸ë¶€ì§–ê¸°!')
        print('ê·¸ëŸ¬ë‚˜ ì•„ë¬´ ì¼ë„ ì¼ì–´ë‚˜ì§€ ì•Šì•˜ë‹¤.. ')
        

    def check_battle(self, target):
        if target.hp <= 0:
            print(f'\n{target.name}ì´(ê°€) ì“°ëŸ¬ì¡ŒìŠµë‹ˆë‹¤! ğŸ’« ')
            self.exp += target.level * 15
            
            print(f'{self.name}ì˜ ìŠ¹ë¦¬!')
            
            if self.exp >= self.level * 100:
                self.exp -= self.level * 100
                self.level += 1
                self.hp = self.level * 20
                print(f'ë ˆë²¨ì—…ì„ í•˜ì—¬ {self.level} ì´ ë˜ì—ˆê³ , í˜„ì¬ ê²½í—˜ì¹˜ëŠ” {self.exp} ì…ë‹ˆë‹¤!')
                
                
    def battle(self, target):
        count = 0
        while self.hp >= 0 and target.hp >= 0:
            print(f'#### ë¼ìš´ë“œ {count} ####')
            print(f'{self.name}ì˜ ì²´ë ¥ : {self.hp}')
            print(f'{target.name}ì˜ ì²´ë ¥ : {target.hp}')
            print('-----------------------------------')
            
            attacker = random.choice([self, target])
            defencer = target if attacker == self else self
        
            skill_number = random.randint(1, 3)
            
            if skill_number == 1:
                attacker.property_skill(defencer)
            elif skill_number == 2:
                attacker.character_skill(defencer)
            else:
                attacker.bark()
            
            count += 1
            print('\n')
        
        self.hp = self.level * 20
        target.hp = target.level * 20
    
        
````

````python
class WaterPokemon(Pokemon):
    property = 'water'
    
    def property_skill(self, target):
        damage = self.level * 3
        target.hp -= damage
        print(f'{self.name}ì˜ ë¬¼ëŒ€í¬! ğŸ’¦ ')
        print(f'{target.name}ì—ê²Œ {damage} ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤!')
        self.check_battle(target)
        
class ThunderPokemon(Pokemon):
    property = 'thunder'
    
    def property_skill(self, target):
        damage = self.level * 4
        target.hp -= damage
        print(f'{self.name}ì˜ ê°ì „ì‹œí‚¤ê¸°')  
        print(f'{target.name}ì—ê²Œ {damage} ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤.')
        self.check_battle(target)
        
class NormalPokemon(Pokemon):
    property = 'normal'
    
    def property_skill(self, target):
        damage = self.level * 4
        target.hp -= damage
        print(f'{self.name}ì˜ ëª¸í†µë°•ì¹˜ê¸°!!! ğŸ‘ŠğŸ»ğŸ‘ŠğŸ»')  
        print(f'{target.name}ì—ê²Œ {damage} ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤.')
        self.check_battle(target)
````
